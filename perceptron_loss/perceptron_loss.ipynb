```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([
    [2, 3],
    [1, 5],
    [2, 4],
    [3, 2],
    [6, 1],
    [7, 2],
    [5, 1]
])

Y = np.array([-1,-1,-1,+1,+1,+1,+1])

W = np.zeros(X.shape[1])
b = 0.0
learning_rate = 0.1
n_epochs = 10
```

```python
for epoch in range(n_epochs):
    total_loss = 0.0
    
    for i in range(len(X)):
        X_i = X[i]
        Y_i = Y[i]
        output = np.dot(W, X_i) + b
        prediction = np.sign(output)
        
        # Calculamos la perdida para la muestra
        loss_i = max(0, -Y_i * output)
        total_loss += loss_i
        
        # Actualizamos si hay error
        if prediction != Y_i:
            W += learning_rate * Y_i * X_i
            b += learning_rate * Y_i
        
    # Imprimimos la perdida total de la época
    print(f'Época {epoch+1}: Perceptron Loss = {total_loss:.4f}')

# Visualzamos la gráfica con la rfonera de desición

def plot_decission_boundary(X, Y, W, b):
    plt.figure(figsize=(8,6))
    
    for i in range(len(X)):
        if Y[i] == -1:
            plt.scatter(X[i][0], X[i][1], color="red", label='Clase -1' if i == 0 else "")
        else:
            plt.scatter(X[i][0], X[i][1], color="blue", label='Clase +1' if i == 3 else "")
    
    x_vals = np.linspace(0,8,100)
    y_vals = -(W[0] * x_vals+b) / W[1]
    plt.plot(x_vals, y_vals, color='black', label='Fronera de desición')
    
    plt.xlabel('x_1 ( característica 1)')
    plt.ylabel('x_2 ( característica 2)')
    plt.title('Perceptrón con visualización de perdida')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_decission_boundary(X, Y, W, b)
```
